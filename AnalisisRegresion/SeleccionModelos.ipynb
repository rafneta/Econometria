## Selecci√≥n de modelos 

```sas
/*LEEMOS LOS DATOS DEL ARCHIVO DE EXCEL
FILENAME REFFILE '/folders/myfolders/AnalisisRegresion/SMSA1.xlsx';
PROC IMPORT DATAFILE=REFFILE
	DBMS=XLSX
	OUT=WORK.datos;
	GETNAMES=YES;
RUN;
quit;

proc glmselect data=work.datos plots=all;
   model y=x1 x2 x3 x4 x5 x6 x7 x8 x9 x10/ selection=backward
                   details=steps select=SL slstay=0.05 slentry=0.05;
   title "Modelo mediante la selecci√≥n backward";
run;
quit;

proc glmselect data=work.datos plots=all;
   model y=x1 x2 x3 x4 x5 x6 x7 x8 x9 x10/ selection=forward
                   details=steps select=SL slstay=0.05 slentry=0.05;
   title "Modelo mediante la selecci√≥n forward";
run;
quit;

proc glmselect data=work.datos plots=all;
   model y=x1 x2 x3 x4 x5 x6 x7 x8 x9 x10/ selection=stepwise
                   details=steps select=SL slstay=0.05 slentry=0.05;
   title "Modelo mediante la selecci√≥n stepwise";
run;
quit;



ods graphics / imagemap=on ;
proc reg data=work.datos plots(only)=(cp);  
model y=x1 x2 x3 x4 x5 x6 x7 x8 x9 x10/ selection=cp rsquare adjrsq ;
title 'Mejor modelo Hocking';
run;
quit;
```
