```sas
/*LEEMOS LOS DATOS DEL ARCHIVO DE EXCEL*/
FILENAME REFFILE '/folders/myfolders/AnalisisRegresion/SMSA1.xlsx';
PROC IMPORT DATAFILE=REFFILE
	DBMS=XLSX
	OUT=WORK.datos;
	GETNAMES=YES;
RUN;
quit;


PROC PRINT DATA = datos;
   TITLE 'Datos Para regresi√≥n';
RUN;
QUIT;

proc contents data=datos;
run;
QUIT;

%let interval= x1 x2 x3 x4 x5 x6 x7 x8 x9 x10; 



proc corr data=work.datos rank plots(only)=scatter(nvar=all ellipse=none);
var x1 x2 x3 x4 x5 x6 x7 x8 x9 x10;
with y;
title 'Correlacion de las variables de entrada con la salida';
run;

ods graphics / imagemap=on ;
proc corr data=work.datos nosimple plots= matrix(nvar=all
histogram ) ;
var y x1 x2 x3 x4 x5 x6 x7 x8 x9 x10;
title 'Correlacion entre todas las variables';
run;
quit;

```
