## Modelo par√°metros significativas

```sas
FILENAME REFFILE '/folders/myfolders/AnalisisRegresion/SMSA1.xlsx';
PROC IMPORT DATAFILE=REFFILE
	DBMS=XLSX
	OUT=WORK.datos;
	GETNAMES=YES;
RUN;
quit;

PROC REG DATA=datos; 
	model y = x2 x5 x6  x10 /noint  VIF SPEC DW clb r cli clm;
	output out=out  predicted=yestimada 
	Residual=residuales 
	lcl=ciPredicIndivi
   ucl=csPredicIndivi
   lclm=ciEsperado
   uclm=csEsperado;


run;
quit;

proc univariate data=out normal; /* Anderson-Darling */
	var residuales;
	title 'Normalidad: Anderson-Darling ';
run;
quit;

data newdata;
merge out datos;
run;
quit;

PROC PRINT DATA =newdata;
RUN;
QUIT;


proc sgplot data=newdata;
   series x=n y=residuales; 
run;
```
